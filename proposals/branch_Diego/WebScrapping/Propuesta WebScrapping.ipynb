```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

import pandas as pd
import numpy as np

from io import StringIO
import os
import time
import requests

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
url     = 'https://coinmarketcap.com/historical/20130428/'
driver  = webdriver.Chrome()        
driver.get( url )
driver.maximize_window()

dataframes_list = []

for index in np.arange( 1, 2808, step = 1 ):
    
    T = driver.find_element( By.XPATH, '//*[@id="__next"]/div[2]/div[2]/div/div[1]/h1').text
    
    time.sleep(1)
    
    table_element = driver.find_element( By.XPATH, '//*[@id="__next"]/div[2]/div[2]/div/div[1]/div[3]/div[1]/div[3]/div/table' )
    table_html    = table_element.get_attribute( 'outerHTML' )
    table_html_io = StringIO( table_html )
    df      = pd.read_html( table_html_io )[ 0 ]

    df_cleaned = df.iloc[:, :].copy()
    df_cleaned.insert(0, 'Time', T)

    time.sleep(2)
    
    # Storing the DataFrame in the list
    dataframes_list.append(df_cleaned)
    
    while T == 'Historical Snapshot - 28 April 2013':
        next_button   = driver.find_element( By.XPATH, '//*[@id="__next"]/div[2]/div[2]/div/div[1]/div[2]/div[2]/a[1]' )
        break
    else:
        next_button   = driver.find_element( By.XPATH, '//*[@id="__next"]/div[2]/div[2]/div/div[1]/div[2]/div[2]/a[2]' )
    
    next_button.click()

    time.sleep( 2 )
    
    print( f'Success at table: { index }' )

# Concatenating all the DataFrames into one, removing the columns we don't want
final_df = pd.concat(dataframes_list, ignore_index=True)
columns_to_drop = ['Rank', 'Unnamed: 9']
final_df = final_df.drop(columns=columns_to_drop, errors='ignore')
```
