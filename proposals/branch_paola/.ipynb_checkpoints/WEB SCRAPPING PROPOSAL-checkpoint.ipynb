```python
# We establish the creation of a folder
path_qaliwarma_data    = './data'

os.makedirs( path_qaliwarma_data, exist_ok = True )
```

## Complete web scrapper code

```python

driver = webdriver.Chrome() 
driver.maximize_window()
for k in range(2014,2016): #The years that exist on the page is since 2009
    driver.get("https://www.transparencia.gob.pe/enlaces/pte_transparencia_enlaces.aspx?id_entidad=14219")
    
    print(k)    
    
    # Selecting the Contrataciones se bienes y servicios button
    
    contrataciones_button = driver.find_element( By.XPATH, '/html/body/section[2]/div/div/div/div/div/div/div[3]/div/ul/li[7]/a')
    contrataciones_button .click()
    
    time.sleep(2)
    
    # Selecting the ordenes de servicio button
    
    ordenes_button = driver.find_element( By.XPATH, '/html/body/section[2]/div/div/div/div/div/div/table/tbody/tr[1]/td[3]/p/a')
    ordenes_button .click()

    time.sleep(1.5)
    
    # Choosing the year from the dropdown (loop by year)
  
    año_select = driver.find_element(By.XPATH, '/html/body/section[2]/div/div/div/div/div/div/form/table/tbody/tr[1]/td[2]/select')
    año_select.click()

    # Using the Select class to interact with the select element
    año_select_obj = Select(año_select)
    año_select_obj.select_by_value(str(k))
       
    time.sleep(2)

    lista_meses = list(range(2, 14))

    # Preparing loop to go for each month
    for i in lista_meses: 
        mes = driver.find_element(By.XPATH, "/html/body/section[2]/div/div/div/div/div/div/form/table/tbody/tr[2]/td[2]/select/option["+str(i)+"]").text # row i, columna 2
        print(mes) 
      
        driver.find_element(By.XPATH, "/html/body/section[2]/div/div/div/div/div/div/form/table/tbody/tr[2]/td[2]/select/option["+str(i)+"]").click()
              
        time.sleep(3)
        # enviar btton
        enviar_button = driver.find_element(By.XPATH, '/html/body/section[2]/div/div/div/div/div/div/form/table/tbody/tr[3]/td/input')
        enviar_button .click()
        #table element
        
        table_element = driver.find_elements(By.XPATH, "//table[@class='table']")

        # Checking if the table contains values, and if so proceed to download
        if table_element:
            table_html = table_element[0].get_attribute('outerHTML')
            table_html_io = StringIO(table_html)
            table_df = pd.read_html(table_html_io)[0]

        #making the first observation the name of the columns
        table_df .columns = table_df .iloc[0]
        table_df  = table_df[1:].reset_index(drop=True)
        # Downloading Excel files ###############################
        # Constructing the file path with the correct string formatting
        table_path = path_qaliwarma_data + f'/contrataciones_{k}_{i}.xlsx'

        table_df.to_excel(table_path, index=False)
       
        time.sleep(3)
    
        # Return to change month
        otro_mes= driver.find_element(By.XPATH, "/html/body/section[2]/div/div/div/div/div/div/form/table/tbody/tr[2]/td[2]/select/option["+str(i)+"]")
        otro_mes.click()
        
        time.sleep(3)
```
